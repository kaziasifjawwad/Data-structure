```python
from prettytable import PrettyTable
#closing bracket case
def extractor(stack,start,end):
  flag=" "
  tempstr=""
  while flag not in start:
    flag=stack.pop()
    # print(a)
    if flag not in start and flag not in end:
      tempstr=tempstr+flag
  return tempstr

#priority of element identifier
def identifyPriority(priority_list,sym , open , close):
  for i in range(len(priority_list)):
    if sym in priority_list[i]:
      return i
    elif sym in open or sym in close:
      return "p"
      break
  return None

def execute(open,close,stack,string,priority_list):
  print("The equation is ==>",string)
  x = PrettyTable()
  x.field_names = ["items","items in stack" , "result"]
  result=""
  string=string.split()


  for item in string:
    priority = identifyPriority(priority_list,item , open , close)
    if item in open:
      stack.append(item)

    elif item in close:
      #print(stack ,"======" , result)
      stack.append(item)
      result=result + extractor(stack,open,close) + " "
    
    elif priority==None:
      #print(stack)
      result=result+ item  +" "
    
    elif len(stack)==0:
      stack.append(item)
    
    else:
      
      lastItem=stack[-1]
      pcurrent = priority
      plast = identifyPriority(priority_list,lastItem , open , close)
      #print(pcurrent,plast , stack)

      if lastItem in open:
        stack.append(item)

      elif pcurrent< plast:
        stack.append(item)
      
      else:
        t = stack.pop()
        result = result + t + " "
        stack.append(item)
    #print(item,"============>",stack,"=========>",result)
    st=""
    for i in stack:
      st=st+i+" "
    x.add_row([item,st,result])

  st=""
  for i in stack[::-1]:
    result=result+i+" "
    st=st+i+" "
  x.add_row(["BLANK","BLANK",result])
  print(x)
  print("Result is ======> ",result)



# ****************************************************** USER INTERFACE ****************************************************
stack=[]

#The priority will be based on the index of the priority_list.
priority_list=[
                "^",  #first priority
                "*/%",
                "+-",
              "< <= >= >",
              "== !=",
              "&&" #second priority
]
open="({["  #do not change it
close=")}]" #do not change it
string = "d >= 4 != [ { ( t == w + 6 ) && ( 4 == 9 / 5 ) } == r * 5 ] != ( 3 + 6 == d )"
execute(open,close,stack,string,priority_list)
```
